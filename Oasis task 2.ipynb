*PIE PLOTS*

*PARTITION ACCORDING TO AREA on "Unemployment in india" dataset*

```python
fig, ax = plt.subplots(figsize =(8, 7))
#labels
m=['Rural','Urban']
#title for chart
ax.set_title("\nPARTITION ACCORDING TO AREA IN \"Unemployment in india\" dataset",color="brown")
# Creating color parameters
colors = ( "blue" ,"magenta")
# Creating explode data
explode = (0.2, 0.0)
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
# Creating plot
wedges, texts, autotexts = ax.pie(df1['Area'].value_counts(),
                                  autopct = "%0.2f",
                                  explode = explode,
                                  labels = m,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 105,
                                  wedgeprops = wp,
                                  textprops = dict(color ="brown"))
# Adding legend
ax.legend(wedges,m,
          title ="AREA",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
 
plt.setp(autotexts, size = 8, weight ="bold")

plt.show()
```

```python
fig, ax = plt.subplots(figsize =(8, 7))
#labels
m=['South', 'Northeast', 'East', 'West', 'North']
#title for chart
ax.set_title("\nPARTITION ACCORDING TO REGION DIRECTIONS",color="brown")
# Creating color parameters
colors = ( "orange", "cyan", "brown","grey", "indigo")
# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0)
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
# Creating plot
wedges, texts, autotexts = ax.pie(df2['Region.1'].value_counts(),
                                  autopct = "%0.2f",
                                  explode = explode,
                                  labels = m,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))
# Adding legend
ax.legend(wedges,m,
          title ="Directions",
          loc ="center left",
          bbox_to_anchor =(1.2, 0, 0.5, 1))
 
plt.setp(autotexts, size = 8, weight ="bold")

plt.show()
```
